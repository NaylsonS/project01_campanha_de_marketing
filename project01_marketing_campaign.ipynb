### 01.01.02 - Exerc√≠cios SQL - Uni√£o de Tabelas

#### 1. Selecionar os n√∫meros do cart√£o de fidelidade dos passeiros, a cidade e o g√™nero, dos passageiros que tem o cart√£o Star de fidelidade, mas nunca realizaram nenhuma viagem de avi√£o ( Dica: ‚Äúloyalty_card‚Äù = ‚ÄúStar‚Äù e ‚Äúdistance‚Äù = 0 )

```python
# consulta dos dados no banco
consulta_atividade = """
  SELECT
    fa.loyalty_number,
    flh.city,
    flh.gender,
    flh.loyalty_card
  FROM
    flight_activity fa LEFT JOIN flight_loyalty_history flh ON ( fa.loyalty_number = flh.loyalty_number )
  WHERE
    flh.loyalty_card = "Star"
      AND fa.distance = 0
"""
df_atividade = pd.read_sql_query( consulta_atividade, conn )

# visualiza os dados
df_atividade.head()
```

#### 2. Selecionar os n√∫meros do cart√£o de fidelidade, o g√™nero e a cidade de todos os passageiros do sexo feminino que moram na cidade de Toronto, fizeram mais de 30 viagens no total e tem o cart√£o de fidelidade do tipo Aurora.


```python
# consulta dos dados no banco
consulta_atividade = """
  SELECT
    fa.loyalty_number,
    fa.total_flights,
    flh.gender,
    flh.city,
    flh.loyalty_card
  FROM
    flight_activity fa LEFT JOIN flight_loyalty_history flh ON ( fa.loyalty_number = flh.loyalty_number )
  WHERE
    flh.gender = 'Female'
      AND flh.city = 'Toronto'
      AND fa.total_flights > 30
      AND flh.loyalty_card = 'Aurora'
"""
df_atividade = pd.read_sql_query( consulta_atividade, conn )

# visualiza os dados
df_atividade.head()
```

#### 3. Selecionar os n√∫meros do cart√£o de fidelidade, o tipo do cart√£o, o genero e os pontos acumulados, dos passageiros com sal√°rio acima de 13200, estado civil como casado e n√≠vel acad√™mico como mestrado e n√∫mero de voos agendados igual ao n√∫mero total de voos.


```python
# consulta dos dados no banco
consulta_atividade = """
  SELECT
    fa.loyalty_number,
    fa.points_accumulated,
    fa.flights_booked,
    fa.total_flights,
    flh.loyalty_card,
    flh.gender,
    flh.marital_status,
    flh.education
  FROM
    flight_activity fa LEFT JOIN flight_loyalty_history flh ON ( fa.loyalty_number = flh.loyalty_number )
  WHERE
    flh.salary > 13200
      AND flh.marital_status = 'Married'
      AND flh.education = 'Master'
      AND ( fa.flights_booked = fa.total_flights )
"""
df_atividade = pd.read_sql_query( consulta_atividade, conn )

# visualiza os dados
df_atividade.head()
```
