#### 1. Selecionar os números do cartão de fidelidade dos passeiros, a cidade e o gênero, dos passageiros que tem o cartão Star de fidelidade, mas nunca realizaram nenhuma viagem de avião ( Dica: “loyalty_card” = “Star” e “distance” = 0 )

```python
# consulta dos dados no banco
consulta_atividade = """
  SELECT
    fa.loyalty_number,
    flh.city,
    flh.gender,
    flh.loyalty_card
  FROM
    flight_activity fa LEFT JOIN flight_loyalty_history flh ON ( fa.loyalty_number = flh.loyalty_number )
  WHERE
    flh.loyalty_card = "Star"
      AND fa.distance = 0
"""
df2_atividade = pd.read_sql_query( consulta_atividade, conn )

# visualiza os dados
df2_atividade.head()
```

#### 2. Selecionar os números do cartão de fidelidade, o gênero e a cidade de todos os passageiros do sexo feminino que moram na cidade de Toronto, fizeram mais de 30 viagens no total e tem o cartão de fidelidade do tipo Aurora.


```python
# consulta dos dados no banco
consulta_atividade = """
  SELECT
    fa.loyalty_number,
    fa.total_flights,
    flh.gender,
    flh.city,
    flh.loyalty_card
  FROM
    flight_activity fa LEFT JOIN flight_loyalty_history flh ON ( fa.loyalty_number = flh.loyalty_number )
  WHERE
    flh.gender = 'Female'
      AND flh.city = 'Toronto'
      AND fa.total_flights > 30
      AND flh.loyalty_card = 'Aurora'
"""
df2_atividade = pd.read_sql_query( consulta_atividade, conn )

# visualiza os dados
df2_atividade.head()
```

#### 3. Selecionar os números do cartão de fidelidade, o tipo do cartão, o genero e os pontos acumulados, dos passageiros com salário acima de 13200, estado civil como casado e nível acadêmico como mestrado e número de voos agendados igual ao número total de voos.


```python
# consulta dos dados no banco
consulta_atividade = """
  SELECT
    fa.loyalty_number,
    fa.points_accumulated,
    fa.flights_booked,
    fa.total_flights,
    flh.loyalty_card,
    flh.gender,
    flh.marital_status,
    flh.education
  FROM
    flight_activity fa LEFT JOIN flight_loyalty_history flh ON ( fa.loyalty_number = flh.loyalty_number )
  WHERE
    flh.salary > 13200
      AND flh.marital_status = 'Married'
      AND flh.education = 'Master'
      AND ( fa.flights_booked = fa.total_flights )
"""
df2_atividade = pd.read_sql_query( consulta_atividade, conn )

# visualiza os dados
df2_atividade.head()
```

# 3.0 - Preparação dos Dados para Treinamento

```python
df2 = df.copy()
```

```python
df2.shape
```

```python
df2.isna().sum()
```

```python
# A princípio vamos utilizar apenas dados numéricos e evitar usar colunas com muitos dados faltantes (cancellation_year, cancellation_month)
# Selecionando apenas colunas com valores númericos
cols = ["year", "month", "flights_booked",
"flights_with_companions", "total_flights",
"distance", "points_accumulated", "salary",
"clv", "loyalty_card"]


df_cols_selected = df2.loc[:, cols]
```

```python
df_cols_selected.isna().sum()
```

```python
# A princípio ficou definido em remover todos os dados que contenham valores nulos
# Removendo linhas que contenham dados faltantes
df_treino = df_cols_selected.dropna()

df_treino.shape
```

# 4.0 - Machine Learning

## 4.1 - Treinando o Algoritmo

```python
from sklearn import tree as tr
```

```python
df_treino.head()
```

```python
# Definindo as variáveis/atributos
X = df_treino.drop( columns='loyalty_card' )
# Definindo a variável resposta
y = df_treino.loc[ :, 'loyalty_card' ]

# Código para definir o algoritmo Decision Tree
modelo = tr.DecisionTreeClassifier( max_depth=5 )

# Código para treinar o algoritmo
modelo_treino = modelo.fit( X, y )
```

```python
# Mostra como ficou o desenho da àrvore
tr.plot_tree( modelo_treino, filled=True )
```

## 4.2 - Fazendo as previsões

```python
# Usando o 'sample' para buscar uma linha aleatória e testar a previsão para o "novo cliente"
X_novo = X.sample()
```

```python
# Verificando a probabilidade do cliente aleatório comprar o plano de fidelidade
previsao = modelo_treino.predict_proba( X_novo )

print( 'Prob - Aurora: {:.2f}% - Nova: {:.2f}% - Star: {:.2f}%'.format( 100*previsao[0][0], 100*previsao[0][1], 100*previsao[0][2] ) )
```
